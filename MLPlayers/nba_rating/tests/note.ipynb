```python
import pandas as pd
from pathlib import Path

# Ajuste ce chemin si besoin pour pointer sur ton dossier « curated »
CURATED = Path("../data/curated")

expected_feats = [
    'efg_pct','ts_pct','stl_mean','blk_mean','tov_mean',
    'pts36','reb36','ast36','stl36','blk36','tov36','pm36'
]

records = []
for path in sorted(CURATED.glob("player_season_*.parquet")):
    season = path.stem.split("_")[-1]
    cols = pd.read_parquet(path).columns.tolist()
    missing = [f for f in expected_feats if f not in cols]
    records.append({"season": season, "missing_features": missing})

df_check = pd.DataFrame(records)

# 1) soit :
df_check

# 2) ou, si tu préfères, :
# from IPython.display import display
# display(df_check)
```
