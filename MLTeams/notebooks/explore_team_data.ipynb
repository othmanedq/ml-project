```python
# 1. Imports
import pandas as pd
import os

# 2. Définir le chemin vers les fichiers
DATA_DIR = "../data/raw"  # Chemin relatif depuis notebooks/

# 3. Charger toutes les saisons disponibles
dfs = []

for season in ["2018-19", "2019-20", "2020-21", "2021-22", "2022-23", "2023-24"]:
    file_path = os.path.join(DATA_DIR, f"team_gamelog_{season}.parquet")
    if os.path.exists(file_path):
        df_season = pd.read_parquet(file_path)
        df_season["season"] = season  # Ajouter une colonne pour savoir de quelle saison vient chaque ligne
        dfs.append(df_season)
    else:
        print(f"⚠️ Fichier manquant pour la saison {season}")

# 4. Concaténer toutes les saisons en un seul DataFrame
if dfs:
    df = pd.concat(dfs, ignore_index=True)
    print(f"✅ Données chargées : {df.shape[0]} lignes et {df.shape[1]} colonnes")
else:
    print("❌ Aucun fichier chargé.")

# 5. Afficher les premières lignes
df.head()
```

```python
# 1. Imports et configuration
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Affichage large pour les DataFrames
pd.set_option('display.max_columns', 30)
sns.set(style="whitegrid")
```

```python
# 2. Chargement des données
# Adapter le chemin si besoin (ici, depuis notebooks/)
df = pd.read_parquet("../data/curated/team_gamelog_all_seasons.parquet")
print(df.shape)
df.head()
```

```python
# 3. Aperçu des saisons et équipes
print("Saisons disponibles :", sorted(df["season"].unique()))
print("Nombre d'équipes différentes :", df["TEAM_NAME"].nunique())
print("Exemples d'équipes :", df["TEAM_NAME"].unique()[:10])
```

```python
# 4. Statistiques descriptives globales
df.describe().T
```

```python
# 5. Nombre de matchs par saison
matchs_par_saison = df.groupby("season")["GAME_ID"].nunique()
print(matchs_par_saison)
matchs_par_saison.plot(kind="bar", title="Nombre de matchs par saison")
plt.ylabel("Nombre de matchs")
plt.show()
```

```python
# 6. Nombre de victoires par équipe (toutes saisons confondues)
victoires = df[df["WL"] == "W"].groupby("TEAM_NAME")["GAME_ID"].count().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=victoires.values, y=victoires.index, palette="crest")
plt.title("Nombre total de victoires par équipe (toutes saisons)")
plt.xlabel("Victoires")
plt.ylabel("Équipe")
plt.show()
```

```python
# 7. Moyenne de points marqués par équipe sur toutes les saisons
pts_mean = df.groupby("TEAM_NAME")["PTS"].mean().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=pts_mean.values, y=pts_mean.index, palette="Blues_r")
plt.title("Moyenne de points marqués par équipe (toutes saisons)")
plt.xlabel("Points moyens")
plt.ylabel("Équipe")
plt.show()
```

```python
# 8. Distribution des scores par match
plt.figure(figsize=(8,5))
sns.histplot(df["PTS"], bins=30, color="skyblue")
plt.title("Distribution des scores par match")
plt.xlabel("Points marqués")
plt.ylabel("Fréquence")
plt.show()
```

```python
# 9. Analyse d'une saison spécifique (exemple : 2022-23)
saison = "2022-23"
df_s = df[df["season"] == saison]
print(f"Nombre de matchs pour la saison {saison} :", df_s.shape[0])

# Top 10 équipes par points moyens
pts_s = df_s.groupby("TEAM_NAME")["PTS"].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(8,5))
sns.barplot(x=pts_s.values, y=pts_s.index, palette="Reds_r")
plt.title(f"Top 10 équipes - Points moyens par match ({saison})")
plt.xlabel("Points moyens")
plt.ylabel("Équipe")
plt.show()
```

```python
# 10. Corrélation entre statistiques d'équipe
corr = df[["PTS", "REB", "AST", "STL", "BLK", "TOV", "PLUS_MINUS"]].corr()
plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Corrélation entre statistiques d'équipe")
plt.show()
```

```python
# 11. Exemple : évolution du score moyen par saison
score_moyen = df.groupby("season")["PTS"].mean()
plt.figure(figsize=(8,5))
score_moyen.plot(marker="o")
plt.title("Évolution du score moyen par saison")
plt.xlabel("Saison")
plt.ylabel("Points moyens")
plt.show()
```

```python
# Moyenne par équipe sur toutes les saisons (rebonds, passes, interceptions, contres)
for stat in ["REB", "AST", "STL", "BLK"]:
    stat_mean = df.groupby("TEAM_NAME")[stat].mean().sort_values(ascending=False)
    plt.figure(figsize=(10,6))
    sns.barplot(x=stat_mean.values, y=stat_mean.index, palette="crest")
    plt.title(f"Moyenne de {stat} par équipe (toutes saisons)")
    plt.xlabel(f"{stat} moyen")
    plt.ylabel("Équipe")
    plt.show()
```

```python
# Créer une colonne "HOME" (1 si l'équipe joue à domicile, 0 sinon)
if "HOME" not in df.columns:
    df["HOME"] = df["MATCHUP"].apply(lambda x: 1 if "vs." in x else 0)

# Moyenne de points à domicile vs extérieur
home_stats_pts = df.groupby("HOME")["PTS"].mean()
plt.bar(["Extérieur", "Domicile"], home_stats_pts.values, color=["#7fa7d6", "#e07a5f"])
plt.title("Points moyens à domicile vs extérieur")
plt.ylabel("Points moyens")
plt.show()

# Victoires à domicile vs extérieur
victoires_home = df[df["WL"] == "W"].groupby("HOME")["GAME_ID"].count()
plt.bar(["Extérieur", "Domicile"], victoires_home.values, color=["#7fa7d6", "#e07a5f"])
plt.title("Nombre de victoires à domicile vs extérieur")
plt.ylabel("Victoires")
plt.show()
```

```python
# Extraire le nom de l'adversaire si la colonne n'existe pas déjà
if "OPPONENT" not in df.columns:
    def get_opponent(row):
        matchup = row["MATCHUP"]
        if "vs." in matchup:
            return matchup.split("vs.")[1].strip()
        elif "@" in matchup:
            return matchup.split("@")[1].strip()
        else:
            return None
    df["OPPONENT"] = df.apply(get_opponent, axis=1)

# Moyenne de points marqués contre chaque adversaire (toutes saisons)
pts_vs_opp = df.groupby("OPPONENT")["PTS"].mean().sort_values(ascending=False)
plt.figure(figsize=(10,6))
sns.barplot(x=pts_vs_opp.values, y=pts_vs_opp.index, palette="mako")
plt.title("Points moyens marqués contre chaque adversaire (toutes saisons)")
plt.xlabel("Points moyens")
plt.ylabel("Adversaire")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="REB", y="PTS", hue="HOME", alpha=0.5)
plt.title("Relation entre rebonds et points (couleur = domicile/extérieur)")
plt.xlabel("Rebonds")
plt.ylabel("Points")
plt.legend(title="Domicile", labels=["Extérieur", "Domicile"])
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Liste des saisons à afficher
saisons = sorted(df["season"].unique())
n = len(saisons)

# Définir la grille (ex: 3 par ligne)
ncols = 3
nrows = int(np.ceil(n / ncols))

fig, axes = plt.subplots(nrows, ncols, figsize=(5*ncols, 5*nrows))

cols = ["PTS", "REB", "AST", "STL", "BLK", "TOV", "PLUS_MINUS"]

for idx, saison_corr in enumerate(saisons):
    row = idx // ncols
    col = idx % ncols
    ax = axes[row, col] if nrows > 1 else axes[col]
    df_s_corr = df[df["season"] == saison_corr]
    corr = df_s_corr[cols].corr()
    sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", ax=ax)
    ax.set_title(f"Saison {saison_corr}")

# Masquer les cases vides si le nombre de saisons n'est pas un multiple de ncols
for idx in range(n, nrows * ncols):
    row = idx // ncols
    col = idx % ncols
    fig.delaxes(axes[row, col] if nrows > 1 else axes[col])

plt.tight_layout()
plt.show()
```
