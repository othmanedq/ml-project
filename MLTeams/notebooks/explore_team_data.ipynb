```python
# 1. Imports
import pandas as pd
import os

# 2. Définir le chemin vers les fichiers
DATA_DIR = "../data/raw"  # Chemin relatif depuis notebooks/

# 3. Charger toutes les saisons disponibles
dfs = []

for season in ["2018-19", "2019-20", "2020-21", "2021-22", "2022-23", "2023-24"]:
    file_path = os.path.join(DATA_DIR, f"team_gamelog_{season}.parquet")
    if os.path.exists(file_path):
        df_season = pd.read_parquet(file_path)
        df_season["season"] = season  # Ajouter une colonne pour savoir de quelle saison vient chaque ligne
        dfs.append(df_season)
    else:
        print(f"⚠️ Fichier manquant pour la saison {season}")

# 4. Concaténer toutes les saisons en un seul DataFrame
if dfs:
    df = pd.concat(dfs, ignore_index=True)
    print(f"✅ Données chargées : {df.shape[0]} lignes et {df.shape[1]} colonnes")
else:
    print("❌ Aucun fichier chargé.")

# 5. Afficher les premières lignes
df.head()
```
